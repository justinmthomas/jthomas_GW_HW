```python
# Dependencies
# Dependencies
import pandas as pd
```

```python
# load CSV
df = pd.read_csv("Resources/2016-FCC-New-Coders-Survey-Data.csv", encoding='iso-8859-1', low_memory=False)
df.head()
```

```python
# Read with pandas--low_memory required to suppress errors about mixed data types
```

```python
# Take only columns 0, 1, 2, 3, 4, 6, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111

columns = [0, 1, 2, 3, 4, 6, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111]
extract_df = df.iloc[:,columns]
extract_df.head()
```

```python
# Change reading 0 and 1 to No and Yes, respectively
extract_df_replace = extract_df.replace({0.0: "No", 1.0:"Yes"})
#extract_df['AttendedBootcamp'] = extract_df['AttendedBootcamp'].apply(lambda x: 0 if x=='no' else 1)
extract_df_replace
```

```python
# Extract rows for only those who attended a bootcamp
attended_bootcamp = extract_df_replace.loc[extract_df_replace["AttendedBootcamp"] == "Yes"]
attended_bootcamp.count()
```

```python
# Create DataFrame of the different boot camps that had a significant number of attendees
significant_attendees = attended_bootcamp.groupby("BootcampName")

significant_attendees.head()

count_by_camp = significant_attendees["AttendedBootcamp"].count() 
count_by_camp

count_by_camp_df = pd.DataFrame({"count_by_camp": count_by_camp})
count_by_camp_df
# Sort in descending order 
count_by_camp_df = count_by_camp_df.sort_values(["count_by_camp"], ascending=False)
count_by_camp_df
```

```python
# Count how many attendees of each bootcamp would recommend the bootcamp
recommend_by_camp = significant_attendees["BootcampRecommend"].count() 
recommend_by_camp

#recommend_bootcamp = pd.DataFrame(recommend_by_camp.groupby("BootcampName")["BootcampRecommend"].sum())
#recommend_bootcamp.reset_index(inplace=True)
```

```python
# Merge the two created data frames on the name of tbe bootcamp

df_merge_bootcamp = pd.merge(recommend_by_camp, count_by_camp_df, on='BootcampName')
```

```python
# Calculate percentage of eac bootcamp's students who are recommenders
df_merge_bootcamp["Percent_Reccomend"] = (df_merge_bootcamp["BootcampRecommend"] / df_merge_bootcamp["count_by_camp"])*100
df_merge_bootcamp.reset_index(inplace=True)
df_merge_bootcamp.head()
df_merge_bootcamp['Percent_Reccomend'] = df_merge_bootcamp['Percent_Reccomend'].map('{:,.2f}%'.format)
df_merge_bootcamp
```

```python
# Export to excel and remove index

df_merge_bootcamp.to_excel("output/thomasbootcampoutput2.xlsx", index=False)
```
